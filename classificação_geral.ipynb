###CodificaÃ§Ã£o BinÃ¡ria (Binary Encoding)

A CodificaÃ§Ã£o BinÃ¡ria (Binary Encoding) Ã© uma tÃ©cnica diferente que converte cada categoria em um nÃºmero binÃ¡rio. PorÃ©m nesse caso mantemos todos os exemplos na mesma coluna.

####Vou fazer a importaÃ§Ã£o do train_test_split para fazer essa divisÃ£o, entÃ£o serÃ¡ from sklearn.model_selection import train_test_split.

Agora vou criar X_treino, X_teste, y_treino e y_teste. Podemos invocar o train_test_split, que vai receber o X, y e random_state = 50. Vou passar tambÃ©m o test_size = 0.3 e o stratify = y.
Utilizamos o random_state = 50 para que essa reproduÃ§Ã£o dessa divisÃ£o seja possÃ­vel, e dividimos os dados sendo 30% para teste e os outros 70% para treino. Utilizamos o stratify para manter a mesma proporÃ§Ã£o dos dados de uma classe, tanto nos dados de treino quanto nos dados de teste.


#####Feita a divisÃ£o, o prÃ³ximo passo Ã© ajustar o modelo de Machine Learning. Para isso, preciso importar o modelo de Machine Learning

Agora precisamos instanciar o modelo, entÃ£o o modelo serÃ¡ igual ao RandomForestClassifier(random_state=2). O random_state = 2 Ã© utilizado porque o Random Forest tambÃ©m tem uma questÃ£o de aleatoriedade nele.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo = RandomForestClassifier(random_state=2)

X = df.drop(['booking_status'],axis=1)
y = df['booking_status']

X_treino,X_teste, y_treino, y_teste = train_test_split(X,y, random_state=50, test_size=0.3,stratify=y)
```

```python
y_teste.shape
```

```python
modelo.fit(X_treino,y_treino)
modelo.score(X_teste,y_teste)
```
