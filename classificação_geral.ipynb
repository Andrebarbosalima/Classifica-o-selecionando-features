## Vídeo 1.3 - Treinando um modelo de ML

###Codificação de Label (Label Encoding)
No caso do Label Encoding podemos ter valores como 0, 1, 2 e assim por diante.

###Codificação One-Hot (One-Hot Encoding)
A Codificação One-Hot é uma técnica usada para representar variáveis categóricas, onde cada categoria é transformada em uma nova coluna binária. Essas colunas binárias têm valores de 0 ou 1, com 1 indicando a presença da categoria e 0 indicando a ausência. Essa técnica é usada para lidar com variáveis categóricas em algoritmos de aprendizado de máquina que exigem entradas numéricas, como árvores de decisão, redes neurais, entre outros.

###Codificação Binária (Binary Encoding)

A Codificação Binária (Binary Encoding) é uma técnica diferente que converte cada categoria em um número binário. Porém nesse caso mantemos todos os exemplos na mesma coluna.

####Vou fazer a importação do train_test_split para fazer essa divisão, então será from sklearn.model_selection import train_test_split.

Agora vou criar X_treino, X_teste, y_treino e y_teste. Podemos invocar o train_test_split, que vai receber o X, y e random_state = 50. Vou passar também o test_size = 0.3 e o stratify = y.
Utilizamos o random_state = 50 para que essa reprodução dessa divisão seja possível, e dividimos os dados sendo 30% para teste e os outros 70% para treino. Utilizamos o stratify para manter a mesma proporção dos dados de uma classe, tanto nos dados de treino quanto nos dados de teste.


#####Feita a divisão, o próximo passo é ajustar o modelo de Machine Learning. Para isso, preciso importar o modelo de Machine Learning

Agora precisamos instanciar o modelo, então o modelo será igual ao RandomForestClassifier(random_state=2). O random_state = 2 é utilizado porque o Random Forest também tem uma questão de aleatoriedade nele.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo = RandomForestClassifier(random_state=2)

X = df.drop(['booking_status'],axis=1)
y = df['booking_status']

X_treino,X_teste, y_treino, y_teste = train_test_split(X,y, random_state=50, test_size=0.3,stratify=y)
```

```python
y_teste.shape
```

```python
modelo.fit(X_treino,y_treino)
modelo.score(X_teste,y_teste)
```
